```python
#dropping the columns which wont help model
traindf=traindf.drop(columns=['PassengerId','Name','Ticket','Cabin'])
```

```python
traindf['Age'] = traindf['Age'].fillna(0)
```

```python
#Creating the dummy variables
traindf =pd.get_dummies(traindf)
```

```python
#we created a list called Age_num where we gonna append the categoricall variable fpr Age 
Age_num=[]
for i in traindf["Age"]:
    if i==0.0:
        Age_num.append(0)
    elif i>0 and i<=10:
        Age_num.append(1)
    elif i>10 and i<= 20:
        Age_num.append(2)
    elif i>20 and i<= 30:
        Age_num.append(3)
    elif i>30 and i<= 40:
        Age_num.append(4)
    elif i>40 and i<= 50:
        Age_num.append(5)
    elif i>50 and i<=60:
        Age_num.append(6)
    elif i>60 and i<=70:
        Age_num.append(7)
    else:
        Age_num.append(8)
traindf['Age']= Age_num
```

```python
#classifying the response variable from dataset
trainData = traindf.drop(columns=['Survived'])
trainResponse = traindf['Survived']
```

```python
trainData
```

```python
#Using train_test_split from the dataset with random_state of 4
x_train, x_test, y_train,y_test = train_test_split(trainData,trainResponse,
                                                   train_size= 0.8, random_state=4)
```

```python
print('Train set:', x_train.shape,y_train.shape)
print("Test set:",x_test.shape,y_test.shape)
```

```python
LR= LogisticRegression(solver='liblinear')
LR.fit(x_train,y_train)
print("Logistic Regression score (test):", LR.score(x_test,y_test))
```
